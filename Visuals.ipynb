```python
import numpy as np
import pandas as pd
import ast
import matplotlib.pyplot as plt
```

```python
question_info = pd.read_csv('/home/vagrant/datacourse/MathQuestionTagging/data/question_info_data_2.csv', index_col=0)
tag_info = pd.read_csv('/home/vagrant/datacourse/MathQuestionTagging/data/tag_info_data_2.csv', index_col=0)
```

```python
count = {}

for i in range( question_info.shape[0] ):
    num_keywords = len( ast.literal_eval( question_info.iloc[i]['keywords'] ) )
    if num_keywords in count:
        count[num_keywords] += 1
    else:
        count[num_keywords] = 1
```

```python
# Aggregegate all questions with 7+ tags
x = []
y = []

for i, val in enumerate(count.values()):
    if i < 7:
        x.append(i)
        y.append(val)
    else:
        y[6] += val
```

```python
index = np.arange(max(x) + 1)

fig, ax = plt.subplots()

plt.title('The Distribution of Number of Keywords per Question')

plt.xticks( index + .25, [1,2,3,4,5,6,'7+'] )
plt.xlim([min(x) - .5, max(x) + 1])

# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_color('gray') 
ax.spines['bottom'].set_color('gray') 

# Hide tick marks
ax.yaxis.set_ticks_position('none')
ax.xaxis.set_ticks_position('none')

plt.bar(x, y, width=.5, color='orange', edgecolor = 'none')
```

```python
top_20_tags = tag_info.sort('count', ascending = False).head(20)

x_vals = top_20_tags['keyword'].values
y_vals = top_20_tags['count'].values
```

```python
index = np.arange(20)

plt.figure(figsize=(10,6))

plt.title('Top 20 question tags')
plt.xlim([-.5, 20.5])
plt.xticks(index+.2, x_vals, rotation=75)

ax = plt.gca()

# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Hide tick marks
ax.yaxis.set_ticks_position('none')
ax.xaxis.set_ticks_position('none')


plt.bar(index, y_vals, color='teal')
```

```python
plt.figure(figsize=(9,7))

plt.title('20 Most Frequently Assigned Keywords', y=1.08)

plt.ylim([-max(index) - 1, min(index) + 1])
plt.yticks(-index, x_vals)

ax = plt.gca()

# Hide tick marks
ax.yaxis.set_ticks_position('none')
ax.xaxis.set_ticks_position('none')

# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

ax.tick_params(labelbottom='off',labeltop='on')

plt.barh(-index, y_vals, align='center', color='teal')
```
